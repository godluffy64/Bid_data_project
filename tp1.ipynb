```python
#17

fig = plt.figure()
axes = fig.add_subplot(111)
axes.set_xlabel("Distance")
axes.set_ylabel("Density")
axes.yaxis.label.set_color('b')
axes.set_ylim(0,2)
axes.set_xlim(0,2)

tab_b = np.array([0,2])
a, = axes.plot(tab_b, tab_b, "b", label="Density")

twin1 = axes.twinx()
twin1.set_ylim(0,4)
tab_rx = np.array([0,1,2])
tab_ry = np.array([0,3,2])
b, = twin1.plot(tab_rx, tab_ry, "r", label="Temperature")
twin1.set_ylabel("Temperature", c='r')

twin2 = axes.twinx()
twin2.set_ylim(1, 65)
tab_vx = np.array([0,1,2])
tab_vy = np.array([50,30,15])
c, = twin2.plot(tab_vx, tab_vy, "g", label="Velocity")
twin2.spines["right"].set_position(("axes", 1.15))
twin2.set_ylabel("Velocity", c='g')

axes.tick_params(axis='y', colors='b')
twin1.tick_params(axis='y', colors='r')
twin2.tick_params(axis='y', colors='g')


axes.legend(handles=[a,b,c], loc='upper right')


fig.show()
```
