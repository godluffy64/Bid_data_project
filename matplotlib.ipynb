```python
import matplotlib.pyplot as plt     #1
import numpy as np

x = np.array([1,2,6,8])
y = np.array([3,8,1,10])

fig = plt.figure()
fig.suptitle('ex1')
axes = fig.add_subplot(111)
axes.plot(x,y)
axes.set_xlim(0,9)
axes.set_ylim(0,11)
fig.show()
```

```python
#2


fig = plt.figure()
fig.suptitle('ex2')
axes = fig.add_subplot(111)
axes.set_xlim(-1,30)
axes.set_ylim(-10,10)
x = np.arange(-1,30,0.01)
y = np.log(x)

axes.plot(x,y)
fig.show()
```

```python
#3


fig = plt.figure()
fig.suptitle('ex3')
axes = fig.add_subplot(111)
axes.set_xlim(-1,13)
axes.set_ylim(-1.25,1.25)

x = np.arange(0,12,0.01)

axes.plot(x,np.sin(x)) #bleu
axes.plot(x,np.cos(x)) #orange
fig.show()
```
